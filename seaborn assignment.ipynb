```python
    #Answer : 1
    
Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for creating attractive and informative statistical graphics. 
It offers a variety of plots and charts for different types of data analysis and visualization.

Here are five common types of plots that you can create using Seaborn and their typical uses:    
    
1.Scatter Plot (sns.scatterplot()):

Use: Scatter plots are used to visualize the relationship between two continuous variables. 
Each point on the plot represents a data point, and you can assess the correlation or distribution of the data.

2.Histogram (sns.histplot()):

Use: Histograms display the distribution of a single variable by dividing the data into bins and showing the frequency or density of values within each bin.
They help identify data patterns and understand the data's underlying distribution.

3.Bar Plot (sns.barplot()):

Use: Bar plots are suitable for visualizing categorical data. They show the relationship between a categorical variable and a continuous variable. 
Commonly used for comparing data across different categories or groups.

4.Box Plot (sns.boxplot()):

Use: Box plots (box-and-whisker plots) display the distribution of data, including measures such as the median, quartiles, and potential outliers. 
They are helpful for identifying variations in data and detecting outliers.

5.Heatmap (sns.heatmap()):

Use: Heatmaps are used to visualize data in a matrix or tabular format. They represent data values using colors, 
making it easy to identify patterns, correlations, or relationships between variables.
```

```python
    #Answer : 2
    
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Create a line plot using Seaborn
plt.figure(figsize=(10, 6))  # Adjust the figure size
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

# Add labels and title
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Line Plot of Signal vs. Timepoint for Different Events and Regions")

# Show the plot
plt.legend(title="Event", loc="upper right")
plt.show()
 
```

```python
    #Answer : 3
    
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Create a box plot for 'age' by 'pclass'
plt.figure(figsize=(10, 6))  # Adjust the figure size
sns.boxplot(x='pclass', y='age', data=titanic)
plt.xlabel("Pclass")
plt.ylabel("Age")
plt.title("Box Plot of Age by Pclass")

# Create a box plot for 'fare' by 'pclass'
plt.figure(figsize=(10, 6))  # Adjust the figure size
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.xlabel("Pclass")
plt.ylabel("Fare")
plt.title("Box Plot of Fare by Pclass")

# Show the plots
plt.tight_layout()
plt.show()
    
```

```python
    #Answer : 4
    
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Create a histogram with hue for 'cut'
plt.figure(figsize=(10, 6))  # Adjust the figure size
sns.histplot(data=diamonds, x='price', hue='cut', kde=True, palette='husl')

# Add labels and title
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Histogram of Price with Cut Categories")

# Show the plot
plt.legend(title="Cut", loc="upper right")
plt.show()
 
```

```python
    #Answer :  5
    
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris = sns.load_dataset("iris")

# Create a pair plot with hue for 'species'
sns.set(style="ticks")
sns.pairplot(data=iris, hue="species", markers=["o", "s", "D"], palette="Set1")

# Show the pair plot
plt.show()
    
```

```python
    #Answer : 6
    
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix for the heatmap
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")

# Create a heatmap
plt.figure(figsize=(10, 6))  # Adjust the figure size
sns.heatmap(flights_pivot, cmap="coolwarm", annot=True, fmt="d", linewidths=0.5)

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Flights Dataset Heatmap")

# Show the heatmap
plt.show()
    
```
